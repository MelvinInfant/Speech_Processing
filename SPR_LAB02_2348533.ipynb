```python
import numpy as np
import matplotlib.pyplot as plt

# Define the signal parameters
fs = 1000  # Sampling frequency
f = 5      # Frequency of the sine wave in Hz
t = np.arange(0, 1, 1/fs)

signal = np.sin(2 * np.pi * f * t)  # Sine wave

plt.figure(figsize=(10, 4))
plt.plot(t, signal)
plt.title('Time-Domain Waveform of 5 Hz Sine Wave')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid()
plt.xlim(0, 1)
plt.show()
```

```python
N = 1024  # Number of points for the DTFT
frequencies = np.linspace(-fs/2, fs/2, N)  # Frequency axis

DTFT = np.fft.fftshift(np.fft.fft(signal, N))  # FFT with shift to center frequency
plt.figure(figsize=(10, 4))
plt.plot(frequencies, np.abs(DTFT))
plt.title('Continuous Frequency Spectrum (DTFT)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid()
plt.xlim(-fs/2, fs/2)
plt.show()
```

```python
M = 512  # Number of points for the DFT
DFT = np.fft.fft(signal, M)  # Compute the DFT of the signal
freq_bins = np.fft.fftfreq(M, d=1/fs)  # Frequency bins

plt.figure(figsize=(10, 4))
plt.stem(freq_bins[:M//2], np.abs(DFT)[:M//2], use_line_collection=True)
plt.title('Discrete Frequency Spectrum (DFT)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid()
plt.xlim(0, fs/2)
plt.show()
```

A composite signal

```python

fs = 1000
t = np.arange(0, 1, 1/fs)

signal1 = 0.5 * np.sin(2 * np.pi * 5 * t)
signal2 = 0.8 * np.sin(2 * np.pi * 20 * t)
composite_signal = signal1 + signal2

plt.figure(figsize=(10, 4))
plt.plot(t, composite_signal)
plt.title('Time-Domain Waveform of Composite Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid()
plt.xlim(0, 1)
plt.show()
```

The Discrete-Time Fourier Transform (DTFT)

```python


N = 1024
DTFT = np.fft.fftshift(np.fft.fft(composite_signal, N))
frequencies = np.linspace(-fs/2, fs/2, N)

plt.figure(figsize=(10, 4))
plt.plot(frequencies, np.abs(DTFT))
plt.title('Continuous Frequency Spectrum (DTFT)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid()
plt.xlim(-fs/2, fs/2)
plt.show()
```

The Discrete Fourier Transform (DFT)


```python

M = 512
DFT = np.fft.fft(composite_signal, M)
freq_bins = np.fft.fftfreq(M, d=1/fs)

plt.figure(figsize=(10, 4))
plt.stem(freq_bins[:M//2], np.abs(DFT)[:M//2], use_line_collection=True)
plt.title('Discrete Frequency Spectrum (DFT)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid()
plt.xlim(0, fs/2)
plt.show()
```

An exponentially decaying signal


```python

fs = 1000
t = np.arange(0, 1, 1/fs)

decay_rate = 5
exponential_signal = np.exp(-decay_rate * t)

plt.figure(figsize=(10, 4))
plt.plot(t, exponential_signal, label='Exponential Decay')
plt.title('Time-Domain Waveform of Exponentially Decaying Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid()
plt.xlim(0, 1)
plt.legend()
plt.show()
```

The Discrete-Time Fourier Transform (DTFT)

```python


N = 1024
DTFT = np.fft.fftshift(np.fft.fft(exponential_signal, N))
frequencies = np.linspace(-fs/2, fs/2, N)

plt.figure(figsize=(10, 4))
plt.plot(frequencies, np.abs(DTFT))
plt.title('Continuous Frequency Spectrum (DTFT)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid()
plt.xlim(-fs/2, fs/2)
plt.show()
```

The Discrete Fourier Transform (DFT)

```python


M = 512
DFT = np.fft.fft(exponential_signal, M)
freq_bins = np.fft.fftfreq(M, d=1/fs)

plt.figure(figsize=(10, 4))
plt.stem(freq_bins[:M//2], np.abs(DFT)[:M//2], use_line_collection=True)
plt.title('Discrete Frequency Spectrum (DFT)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid()
plt.xlim(0, fs/2)
plt.show()
```

**Analysis**:
The exponentially decaying signal shows a rapid decrease in amplitude over time, resulting in a frequency-domain representation with a peak at low frequencies.This indicates that most of the signal's energy is concentrated at lower frequencies, which is typical for decaying signals.

 Both plots show that the signal contains primarily low-frequency components, indicated by the sharp peak at or near 0 Hz.
